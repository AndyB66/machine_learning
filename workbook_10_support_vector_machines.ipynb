```
import numpy as np
from sklearn.svm import SVC # "Support vector classifier"

model = SVC(kernel='linear', C=1E5, max_iter=10)

def fit_model(X, y):
  model.fit(X, y)

def predict_result(X):
  return model.predict(X)

def decision_function(X):
  return model.decision_function(X)

def support_vectors():
  return model.support_vectors_
```

```
import numpy as np
import scipy.spatial.distance as dist


class Linear(object):
    def __call__(self, x, y):
        return np.dot(x, y.T)

    def __repr__(self):
        return "Linear kernel"


class Poly(object):
    def __init__(self, degree=2):
        self.degree = degree

    def __call__(self, x, y):
        return np.dot(x, y.T) ** self.degree

    def __repr__(self):
        return "Poly kernel"


class RBF(object):
    def __init__(self, gamma=0.1):
        self.gamma = gamma

    def __call__(self, x, y):
        x = np.atleast_2d(x)
        y = np.atleast_2d(y)
        return np.exp(-self.gamma * dist.cdist(x, y) ** 2).flatten()

    def __repr__(self):
        return "RBF kernel"
```

**SVM** is basically a binary linear classifier. Classes are called +1 and -1 and defined as
$$
\text{class} = \text{sign}\left(\vec{x} \cdot \vec{w} + b\right).
$$
I will call 
$$
\vec{x} \cdot \vec{w} + b
$$
a confidence function -- the more positive it is, the more confident is the classifier that this is +1, same true for -1.

```
import numpy as np

class SVM:
  def __init__(self, kernel='linear', C=10000.0, max_iter=100000, degree=3):
    self.kernel = {'linear': lambda x,y: np.dot(x, y.T),
                   'poly'  : lambda x,y: np.dot(x, y.T)**degree}[kernel]
    self.C = C
    self.max_iter = max_iter

  def get_vector_k(self, idxI, idxJ):
    l = self.lambdas.copy()
    l[[idxI, idxJ]] = [0, 0]
    return 2 * np.array([1 - np.sum(self.K[idxI]*l), 1 - np.sum(self.K[idxJ]*l)])

  def restrict_to_square(self, t, v0, u):
    if (v0 + t*u)[1] < 0: t = -v0[1]/u[1]
    if (v0 + t*u)[1] > self.C: t = (self.C - v0[1])/u[1]
    if (v0 + t*u)[0] < 0: t = -v0[0]/u[0]
    if (v0 + t*u)[0] > self.C: t = (self.C - v0[0])/u[0]
    return t

  def fit(self, X, y):
    self.X = X.copy() # store for decision function
    self.y = y.copy() * 2 - 1 # store for decision function
    self.lambdas = np.zeros_like(self.y, dtype=float) # dual variables
    self.K = np.array([[y1*y2*self.kernel(x1,x2) for x1,y1 in zip(X,y)] for x2,y2 in zip(X,y)]) # get K matrix

    for _ in range(self.max_iter):
      for idxI in range(len(self.lambdas)):
        idxJ = idxI
        while idxJ == idxI: idxJ = np.random.randint(0, len(self.lambdas));

        M = self.K[[[idxI,idxI],[idxJ,idxJ]], [[idxI,idxJ],[idxI,idxJ]]] # get M matrix
        k = self.get_vector_k(idxI, idxJ) # get k vector
        v0 = self.lambdas[[idxI, idxJ]] # initial lambdas
        u = np.array([-self.y[idxJ], self.y[idxI]]) # lambda line direction
        t_max = (np.dot(k, u) - np.dot(np.dot(M, u), v0) - np.dot(np.dot(M, v0), u)) / (2 * np.dot(np.dot(M, u), u))
        self.lambdas[[idxI, idxJ]] = v0 + u * self.restrict_to_square(t_max, v0, u)

    idx, = np.nonzero(self.lambdas > 1E-15)
    self.b = np.sum((1.0 - np.sum(self.K[idx] * self.lambdas, axis=1)) * self.y[idx]) / len(idx)

  def decision_function(self, X):
    return np.sum(self.kernel(X, self.X) * self.y * self.lambdas, axis=1) + self.b

######### TESTS ############
from sklearn.svm import SVC # "Support vector classifier"

svm_theirs = SVC(kernel='linear', C=1E5, max_iter=60, degree=3)
svm_my     = SVM(kernel='linear', C=1E5, max_iter=60, degree=3)


import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
from sklearn.datasets import make_blobs
from matplotlib.colors import ListedColormap

X, y = make_blobs(n_samples=50, centers=2, random_state=0, cluster_std=0.8)
xlim = [np.min(X[:, 0]), np.max(X[:, 0])]
ylim = [np.min(X[:, 1]), np.max(X[:, 1])]
xx, yy = np.meshgrid(np.linspace(*xlim, num=700), np.linspace(*ylim, num=700))
rgb=np.array([[210, 0, 0], [0, 0, 150]])/255.0

svm_theirs.fit(X, y)
Z_theirs = svm_theirs.decision_function(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)

#y = y * 2 - 1
svm_my.fit(X, y)
Z_my = svm_my.decision_function(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)

fig, axs = plt.subplots(nrows=1,ncols=2,figsize=(12,4))

plt.axes(axs[0])
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='autumn')
plt.contour(xx, yy, Z_my, colors='k', levels=[-1, 0, 1], alpha=0.5, linestyles=['--', '-', '--'])
plt.contourf(xx, yy, np.sign(Z_my.reshape(xx.shape)), alpha=0.3, levels=2, cmap=ListedColormap(rgb), zorder=1)

plt.axes(axs[1])
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='autumn')
plt.contour(xx, yy, Z_theirs, colors='k', levels=[-1, 0, 1], alpha=0.5, linestyles=['--', '-', '--'])
plt.contourf(xx, yy, np.sign(Z_theirs.reshape(xx.shape)), alpha=0.3, levels=2, cmap=ListedColormap(rgb), zorder=1)
```

```
import numpy as np

class SVM:
  def __init__(self, kernel='linear', C=10000.0, max_iter=100000, degree=3, gamma=1):
    self.kernel = {'linear': lambda x,y: np.dot(x, y.T),
                   'poly'  : lambda x,y: np.dot(x, y.T)**degree,
                   'rbf'   : lambda x,y: np.exp(-gamma*np.sum(np.squeeze((np.atleast_2d(y) - np.atleast_2d(x)[:,np.newaxis]))**2, axis=-1))}[kernel]
    self.C = C
    self.max_iter = max_iter

  def restrict_to_square(self, t, v0, u):
    if (v0 + t*u)[1] < 0: t = -v0[1]/u[1]
    if (v0 + t*u)[1] > self.C: t = (self.C - v0[1])/u[1]
    if (v0 + t*u)[0] < 0: t = -v0[0]/u[0]
    if (v0 + t*u)[0] > self.C: t = (self.C - v0[0])/u[0]
    return t

  def fit(self, X, y):
    self.X = X.copy() # store for decision function
    self.y = y * 2 - 1 # store for decision function
    self.lambdas = np.zeros_like(self.y, dtype=float) # dual variables
    self.K = np.array([[y1*y2*self.kernel(x1,x2) for x1,y1 in zip(self.X,self.y)] for x2,y2 in zip(self.X,self.y)]) # get K matrix
    #K1 = self.y * np.array([[self.kernel(x1,x2) for x1 in self.X] for x2 in self.X]) * self.y[np.newaxis,]
    #print(self.K - K1)
    
    for _ in range(self.max_iter):
      for idxI in range(len(self.lambdas)):
        idxJ = idxI
        
        while idxJ == idxI: idxJ = np.random.randint(0, len(self.lambdas));
        M = self.K[[[idxI,idxI],[idxJ,idxJ]], [[idxI,idxJ],[idxI,idxJ]]] # get M matrix
        v0 = self.lambdas[[idxI, idxJ]] # initial lambdas
        k = 2 * (1 - np.sum(self.lambdas * self.K[[idxI, idxJ]], axis=1) + np.dot(M, v0))
        u = np.array([-self.y[idxJ], self.y[idxI]]) # lambda line direction
        t_max = (np.dot(k, u) - np.dot(np.dot(M, u), v0) - np.dot(np.dot(M, v0), u)) / (2 * np.dot(np.dot(M, u), u))
        self.lambdas[[idxI, idxJ]] = v0 + u * self.restrict_to_square(t_max, v0, u)
    
    idx, = np.nonzero(self.lambdas > 1E-15)
    self.b = np.sum((1.0 - np.sum(self.K[idx] * self.lambdas, axis=1)) * self.y[idx]) / len(idx)
  
  def decision_function(self, X):
    return np.sum(self.kernel(X, self.X) * self.y * self.lambdas, axis=1) + self.b

######### TESTS ############
from sklearn.svm import SVC # "Support vector classifier"
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
from sklearn.datasets import make_blobs, make_circles
from matplotlib.colors import ListedColormap

X, y = make_blobs(n_samples=50, centers=2, random_state=0, cluster_std=1.3)
#X, y = make_circles(100, factor=.1, noise=.1)

def test_plot(X, y, svm_model, axes, title):
  plt.axes(axes)
  xlim = [np.min(X[:, 0]), np.max(X[:, 0])]
  ylim = [np.min(X[:, 1]), np.max(X[:, 1])]
  xx, yy = np.meshgrid(np.linspace(*xlim, num=700), np.linspace(*ylim, num=700))
  rgb=np.array([[210, 0, 0], [0, 0, 150]])/255.0
  
  svm_model.fit(X, y)
  z_model = svm_model.decision_function(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
  
  plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='autumn')
  plt.contour(xx, yy, z_model, colors='k', levels=[-1, 0, 1], alpha=0.5, linestyles=['--', '-', '--'])
  plt.contourf(xx, yy, np.sign(z_model.reshape(xx.shape)), alpha=0.3, levels=2, cmap=ListedColormap(rgb), zorder=1)
  plt.title(title)

fig, axs = plt.subplots(nrows=1,ncols=2,figsize=(12,4))
test_plot(X, y, SVM(kernel='rbf', C=1E5, max_iter=60, degree=3, gamma=1), axs[0], 'OUR ALGORITHM')
test_plot(X, y, SVC(kernel='rbf', C=1E5, max_iter=60, degree=3, gamma=1), axs[1], 'sklearn.svm.SVC')
```
